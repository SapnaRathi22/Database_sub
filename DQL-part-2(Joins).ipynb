### **Retrieve all rows where there is a matching record in both the facts and cities tables, using the id column in the facts table and the facts_id column in the cities table as the join condition.**

```python
## inner join
q = '''select * from facts f
    inner join cities c
    on
    f.id = c.facts_id

'''
sql_query(q)
```

### **Show the country name and city name for entries where the country is 'Pakistan'**

```python
q = '''select f.name Country, c.name city from facts f
    inner join  cities c
    on
    f.id = c.facts_id
    where f.name == "Pakistan"
    
'''
sql_query(q)
```

### **find countries that do not have any cities listed in the cities table. The query should return the country name and city name.**

```python
##left join
q = '''select f.name Country, c.name city from facts f
    left join  cities c
    on
    f.id = c.facts_id
    where c.name is null
    
'''
sql_query(q)
```

```python
##right join
q = '''select f.name Country, c.name city from facts f
    right join  cities c
    on
    f.id = c.facts_id
    where c.name is null
    
'''
sql_query(q)
```

```python
#joins-- right join
q = ''' select f.name country, c.name as city_name, c.population as city_polpulation  from facts f right join cities c  on
          f.id = c.facts_id
          order by  c.population desc limit 3
     '''
sql_query(q)
```

```python

## full join
q = '''select * from facts f
    full join cities c
    on
    f.id = c.facts_id

'''
sql_query(q)
```
