```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('database.sqlite')
conn
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
q= '''SELECT name FROM sqlite_master'''
sql_query(q)
```

# Select Statement

```python
#Select All Columns Using * 
q= '''SELECT * FROM Salaries'''
sql_query(q)
```

```python
#Select Specific Columns
q= '''SELECT ID, employeeName , Jobtitle  FROM Salaries'''
sql_query(q)
```

```python
#Select Columns with Aliases
q= '''SELECT S.ID, S.employeeName , S.Jobtitle  FROM Salaries S'''
sql_query(q)
```

```python
#Select Columns with Aliases
q= '''SELECT ID, employeeName as name , Jobtitle job  FROM Salaries '''
sql_query(q)
```

```python
#Select Distinct Columns
q= '''SELECT DISTINCT Jobtitle job_types FROM Salaries '''
sql_query(q)
```

```python
# Select Columns Using Expressions or Calculations
q= '''SELECT totalpay+totalpaybenefits as salary  FROM Salaries '''
sql_query(q)
```

```python
# Select Columns Using Expressions or Calculations
q= '''SELECT  Employeename||" "||jobtitle as name_and_job  FROM Salaries '''
sql_query(q)
```

```python
#Select Distinct Columns
q= '''SELECT DISTINCT status
FROM Salaries '''
sql_query(q)
```

```python
#Select Columns Using CASE Statements 
q = '''SELECT EmployeeName,
       CASE
         WHEN Status = 'PT' THEN 'High' 
         WHEN Status = 'FT' THEN 'Low'
         ELSE 'Unknown'
       END AS Category
FROM Salaries; '''
sql_query(q)
```

# Where Clause

```python
q= ''' select EmployeeName as name , jobtitle from Salaries 
   where jobtitle = "GENERAL MANAGER-METROPOLITAN TRANSIT AUTHORITY"  '''
sql_query(q)
```

```python
q= ''' select EmployeeName as name , jobtitle, totalpay , Basepay from Salaries 
   where Basepay >= 134401 and totalpay >=  332343.61	 '''
sql_query(q)
```

```python
# IN
q= ''' select EmployeeName as name , jobtitle From  Salaries 
   where jobtitle IN ('Custodian','CAPTAIN III (POLICE DEPARTMENT)', 'WIRE ROPE CABLE MAINTENANCE MECHANIC')	 '''
sql_query(q)
```

```python
# NOT IN
q= ''' select EmployeeName as name , jobtitle From  Salaries 
   where jobtitle NOT IN ('Custodian','CAPTAIN III (POLICE DEPARTMENT)', 'WIRE ROPE CABLE MAINTENANCE MECHANIC')	 '''
sql_query(q)
```

```python
#Between
q= ''' select EmployeeName as name , jobtitle , Basepay from Salaries 
       where Basepay between 134401 and 332343.61	 '''
sql_query(q)
```

```python
#IS NULL
q= ''' select EmployeeName as name , jobtitle  from Salaries 
       where Jobtitle is NULL	 '''
sql_query(q)
```

```python
#IS NULL
q= ''' select EmployeeName as name , jobtitle  from Salaries 
       where Jobtitle is not NULL	 '''
sql_query(q)
```

# ORDER BY

```python
q= ''' select  EmployeeName as name , jobtitle , totalpay from Salaries 
       order by totalpay
    '''
sql_query(q)       
```

```python
q= ''' select  EmployeeName as name , jobtitle , totalpay from Salaries 
       where totalpay > 457433
       order by totalpay DESC
    '''
sql_query(q)     
```

```python
q= ''' select  EmployeeName as name , jobtitle , totalpay from Salaries 
       where totalpay > 457433
       order by 2 DESC
    '''
sql_query(q)     
```

# Group By

```python
q= ''' select  count(EmployeeName), jobtitle  from Salaries 
       Group by jobtitle
    
    '''
sql_query(q)   
```

```python
q= ''' select  count(EmployeeName), jobtitle ,  totalpay from Salaries 
       where totalpay> 77643
       Group by jobtitle
    
    '''
sql_query(q)   
```

```python
q= ''' select  jobtitle , max( totalpay) as max_salary  , min(totalpay) as min_salary  from Salaries 
       Group by jobtitle 
    '''
sql_query(q)  
```

```python
q= ''' select  jobtitle , max( totalpay) as max_salary  , min(totalpay) as min_salary ,  year  from Salaries 
       Group by jobtitle , year 
    '''
sql_query(q)  
```

```python
q= ''' select EmployeeName , jobtitle , max(totalpay) from Salaries 
       Group by jobtitle
       order by jobtitle desc 
    '''
sql_query(q)  
```

# limit

```python
q= ''' select EmployeeName , jobtitle , max(totalpay) from Salaries 
       Group by jobtitle
       order by jobtitle desc 
       limit 10
    '''
sql_query(q)  
```

# Having 

```python
q= ''' select  count(EmployeeName), jobtitle  from Salaries 
       Group by jobtitle
       having  count(EmployeeName)>10
       order by  count(EmployeeName) desc
    '''
sql_query(q)   
```
